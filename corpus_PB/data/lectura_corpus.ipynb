```python
import pandas as pd

# 1. Cargar el CSV (Asegúrate de que el archivo esté en la misma carpeta que el script)
# Si tu Excel lo guardó con punto y coma, cambia sep=',' por sep=';'
df = pd.read_csv("pb_reference.csv", sep=',') 

print("=" * 80)
print(f"📊 REPORTE DEL CORPUS UPV-EARTH | Total de PBs: {len(df)}")
print("=" * 80)

# 2. Recorrer y mostrar cada fila con un formato humano y limpio
for index, row in df.iterrows():
    # Cabecera del PB
    print(f"\n🌍 [{row['pb_code']}] {str(row['pb_name']).upper()}")
    print(f"   🔹 Aliases: {row['aliases']}")
    print(f"   📖 Definición Corta: {row['short_definition']}")
    print(f"   ⚙️ Variables de Control: {row['control_variables']}")
    print("\n   --- PALABRAS CLAVE (EL MOTOR) ---")
    print(f"   🔑 Core Keywords:     {row['core_keywords']}")
    print(f"   ➕ Extended Keywords: {row['extended_keywords']}")
    print(f"   🛠️ Applied UPV:       {row['applied_keywords_upv']}")
    print(f"   📊 Method Contextual: {row['method_keywords_contextual']}")
    print(f"   🏭 Sector Keywords:   {row['sector_keywords']}")
    print("\n   --- LÓGICA Y REGLAS (EL CEREBRO) ---")
    print(f"   💡 Frases de Ejemplo:   {row['example_phrases']}")
    print(f"   🧠 Lógica de Activación:{row['activation_logic']}")
    print(f"   ⛔ Notas de Exclusión:  {row['exclusion_notes']}")
    print("\n   --- TRAZABILIDAD ---")
    print(f"   📚 Evidencia en Papers: {row['evidence_from_project_papers']}")
    print(f"   🔗 Fuente original:     {row['source_basis']}")
    print("-" * 80)
```
